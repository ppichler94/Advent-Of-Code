```python
def is_valid_a(id: int) -> bool:
    str_id = str(id)
    return str_id[:len(str_id)//2] != str_id[len(str_id)//2:]
```

```python
def invalid_ids(id_range: str, is_valid: Callable[[int], bool]) -> list[int]:
    start, end = id_range.split('-')
    invalid = []
    for i in range(int(start), int(end)+1):
        if not is_valid(i):
            invalid.append(i)
    return invalid
```

```python
def part_a(input: str, is_valid: Callable[[int], bool]):
    ranges = input.split(',')
    invalid_ranges_id = [invalid_ids(it, is_valid) for it in ranges]
    invalid = [it for it1 in invalid_ranges_id for it in it1]
    return sum(invalid)
```

```python
result = part_a(data, is_valid_a)
submit(result, part="a", day=2, year=2025)
```

```python
def find_rotation(id: int) -> int:
    str_id = str(id)
    for i in range(1, len(str_id)):
        if str_id[:i] == str_id[-i:]:
            return i
    return 0

def is_valid_b(id: int) -> bool:
    return find_rotation(id) == 0
```
