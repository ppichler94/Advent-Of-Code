```python
from aocd import get_puzzle, submit
import math
puzzle = get_puzzle(day=1, year=2025)
data = puzzle.input_data
```

```python
example_data = "L68\nL30\nR48\nL5\nR60\nL55\nL1\nL99\nR14\nL82\n"
```

```python
def part_a(data):
    dial = 50
    password = 0
    rotations = data.splitlines()
    for rotation in rotations:
        direction, steps = rotation[0], int(rotation[1:])
        if direction == 'L':
            dial = (dial - steps) % 100
        else:
            dial = (dial + steps) % 100
        if dial == 0:
            password += 1
    return password
    
```

```python
part_a(example_data)
```

```python
result = part_a(data)
submit(result, part="a", day=1, year=2025)
```

```python
def part_b(data):
    dial = 50
    password = 0
    rotations = data.splitlines()
    for rotation in rotations:
        direction, steps = rotation[0], int(rotation[1:])
        if direction == 'L':
            k0 = dial % 100
        else:
            k0 = (100 - dial) % 100

        if k0 == 0:
            k0 = 100

        if steps < k0:
            zero_crossings = 0
        else:
            zero_crossings = 1 + (steps - k0) // 100

        # apply rotation
        if direction == 'L':
            dial = (dial - steps) % 100
        else:
            dial = (dial + steps) % 100

        password += zero_crossings
    return password
    
```

For part B, we have to find the total number of zero crossings.
This can be done in two steps:
- First, smallest positive step `k0` at which the dial hits `0`
  - moving right: `k0 = (100 - dial) % 100`
  - moving left: `k0 = dial % 100`
- Second, find the number of rotations (`clicks // 100`)

```python
part_b(example_data)
```

```python
result = part_b(data)
submit(result, part="b", day=1, year=2025)
```
